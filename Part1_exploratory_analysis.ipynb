**Grouping correlated features - using Pearson correlation**

```python
# Create correlation table
corr = diab[diab_cols].corr()
corr
```

```python
# Function to pass through dataset and an amendable threshold value
def correlation(data, threshold):
    col_corr = set() # Set to hold names of correlated cols
    for i in range(len(corr.columns)):
        for j in range(i):
            if abs(corr.iloc[i, j]) > threshold:
                colname = corr.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
# Pass arguments for the data input and correlation threshold
data = diab[diab_cols]
corr_features = correlation(data, 0.7)
len(set(corr_features))
```

```python
# View features that would be considered to be highly correlated
set(corr_features)
```
